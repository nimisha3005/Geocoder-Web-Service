```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure,output_file,show
from bokeh.models.annotations import Title

start=datetime.datetime(2017,11,25)
end=datetime.datetime(2018,12,27)

df=data.DataReader(name="AAPL",data_source="yahoo",start=start,end=end)
df
```

```python
def inc_dec(c, o):
    if c > o:
        value="Increase"
    elif c < o:
        value="Decrease"
    else:
        value="Equal"
    return value

df["Status"]=[inc_dec(c,o) for c,o in zip(df.Close,df.Open)]
```

```python
df["Middle"]=(df.Open+df.Close)/2
df["Height"]=abs(df.Open-df.Close)
```

```python
p=figure(x_axis_type='datetime', width=1000, height=500)
t=Title()
t.text="Candlestick Chart"
p.title=t

hours_12=12*60*60*1000

p.segment(df.index, df.High, df.index, df.Low, line_color="black")

p.rect(df.index[df.Status=="Increase"],df.Middle[df.Status=="Increase"],hours_12,
       df.Height[df.Status=="Increase"],fill_color="#CCFFFF",line_color="black")

p.rect(df.index[df.Status=="Decrease"],df.Middle[df.Status=="Decrease"],hours_12,
       df.Height[df.Status=="Decrease"],fill_color="red",line_color="black")

output_file("CS.html")
show(p)
```
